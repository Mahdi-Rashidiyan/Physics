```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Calculate motion using Lorentz force
position = np.zeros((steps, 3))
velocity = np.zeros((steps, 3))
velocity[0] = v0
```

```python
# Plot trajectory
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(position[:,0], position[:,1], position[:,2])
ax.set_title("Charged Particle Motion in a Magnetic Field")
ax.set_xlabel("X (m)")
ax.set_ylabel("Y (m)")
ax.set_zlabel("Z (m)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Generate simulated particle energy data
np.random.seed(42)
data = {
    "energy": np.concatenate([
        np.random.normal(100, 10, 500),   # Cosmic rays
        np.random.normal(200, 30, 300),   # High-energy particles
        np.random.normal(50, 5, 200)      # Background noise
    ])
}
df = pd.DataFrame(data)

# Analyze data
plt.figure(figsize=(10, 6))
plt.hist(df['energy'], bins=50, edgecolor='black', alpha=0.7)
plt.title("Energy Distribution of Detected Particles")
plt.xlabel("Energy (MeV)")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```
